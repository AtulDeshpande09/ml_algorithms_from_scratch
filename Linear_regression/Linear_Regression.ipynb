```python
def gradient_descent(x:list , y:list , learning_rate:float=0.001, iterations:int=1000):

    current_weight , current_bias = 0, 0

    n = len(x)
    
    # iterating the main loop over given number of iterations
    for _ in range(iterations):

        #initialise weight and bias derivatives as 0
        weight_derivative = 0
        bias_derivative = 0

        for i in range(n):
            
            # calculating predicted value of y based on formula
            y_predicted = ( current_weight*x[i] + current_bias)

            # Gradients of weight and bias
            weight_derivative += (-2*x[i]/n)*(y[i] - y_predicted)
            bias_derivative += (-2/n)*(y[i] - y_predicted)

        # changing weight and bias after calculating summation of weight and bias derivatives
        # Using minus sign since gradient gives direction to highest point
        # We want to minimise it Hence moving in opposite direction
        current_weight -= learning_rate*weight_derivative
        current_bias -= learning_rate*bias_derivative

    return current_weight , current_bias
```

```python
class linear_regression:
    def __init__(self):
        self.weight = 0
        self.bias = 0
        
    def train(self,x:list , y:list ):
        self.weight , self.bias = gradient_descent(x,y)

    def predict(value:list)->list:
        # Here we will use the Equation of linear Equation
        # here value is the datapoints for which we want to predict
        # y_predict = self.weight*value + self.bias - for single datapoint
        # since we will take list as input we need to modify this

        # First create a zero list to store and return predictions

        y_predict:list = [0]*len(value)

        for i in range(len(value)):
            # Update values of y_predict based on equation
            y_predict[i] = self.weight*value[i] + self.bias

        return y_predict

    def accuracy(y_test:list ,y_predict:list):
        accuracy = [0,0] # List will give MSE and MAE values
        
        mse , mae = 0 , 0
        
        for i in range(len(y_test)):
            mse += 1/n*(y_test[i]-y_predict[i])**2
            mae += 1/n*(y_test[i]-y_predict[i])

        accuracy[0] = mse
        accuracy[1] = mae
        
        print(f"Mean Squared Error : {self.accuracy[0]}")
        print(f"Mean Absolute Error : {self.accuracy[1]}")

        return accuracy
```
