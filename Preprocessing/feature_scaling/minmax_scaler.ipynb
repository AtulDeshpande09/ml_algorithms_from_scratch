```python
class MinMaxScaler:

    def __init__(self):
        self.x_min = None
        self.x_max = None

    def fit(self,data):
        """
        Finds the max and min values from the column
        Parameters:
        data : Array-like data structure 
        """
        self.x_min = min(data)
        self.x_max = max(data)

    def fit_transform(self,data):
        """
        Scales data from range 0 to 1
        Parameters:
        data : Array-like data structure
        Return :
        list : scaled data
        """
        #call fit method to ensure min and max value are correctly updated according to data
        self.fit(data)
        # Min-max scaling
        x_scaled = [(x - self.x_min)/(self.x_max - self.x_min) for x in data]

        return x_scaled
    
    def inverse_transform(self,scaled_data):
        """
        Reverts scaled data to original
        
        Parameters:
        scaled_data : Array-like data structure
        
        Return :
        list : original data
        """

        # Reverting data into original format
        x_original = [m*(self.x_max - self.x_min)+self.x_min for m in scaled_data]

        return x_original
```

```python
data = [4.2 , 4.8 , 5.1 , 4, 6.1 ,6.0, 5.5 , 4.6 , 5.4, 6.5,4.4 , 5.8 , 5.3,4.5,7,2.4,3,7.5,5,6.8,2.7]
scaler = MinMaxScaler()

scaled_data = scaler.fit_transform(data)

print(scaled_data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig = plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Original Data")
sns.scatterplot(x=data,y=[i for i in range(len(data))])
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.subplot(1,2,2)
plt.title("Scaled Data ( using minmax scaling )")
sns.scatterplot(x=scaled_data,y=[i for i in range(len(scaled_data))])
plt.xlabel("X axis")
plt.ylabel("Y axis")
```
